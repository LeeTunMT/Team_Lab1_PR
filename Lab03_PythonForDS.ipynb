# PHẦN 02: NAIVE BAYES CLASSIFICATION

```python
data_new = data[(data.fruit_name=='apple') | (data.fruit_name=='lemon')]
data_new
```

## GaussianNB model for new data

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
X=data_new[['mass','width','height','color_score']]
Y=data_new['fruit_name']


#Chia tập dữ liệu trên dataframe mới thành 2 tập train và test với 30% dữ liệu test
X_train, X_test, Y_train, Y_test =train_test_split(X,Y,test_size=0.3,random_state=42)
```

> Ở đây em không sử dụng đặc trưng `fruit_subtype` bởi vì:
 * `fruit_subtype` không phải đặc trưng độc lập.
 * `fruit_name` và `fruit_subtype` có quan hệ trực tiếp, sử dụng cả hai hàm đặc trưng sẽ gây ra **đa cộng tuyến** - làm dữ liệu chứa thông tin trùng lặp => Giảm khả năng tổng quát hóa mô hình

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn import metrics


gnb=GaussianNB()
gnb.fit(X_train,Y_train)


Y_pred=gnb.predict(X_test)
print("GNB model accuracy(in %): ",metrics.accuracy_score(Y_test,Y_pred)*100)
```
